```python
prices = pd.read_csv('../data/sac_re_cat.csv')
```

```python
prices.head()
```

```python
prices.zip = prices.zip.astype('category')
```

```python
prices.groupby('zip').zip.count()
```

```python
prices.groupby('zip').zip.count().head()
```

```python
prices.groupby('zip').zip.transform('count').head()
```

```python
transformed = prices.groupby('zip').zip.transform('count')
```

```python
prices.groupby('zip').zip.transform('count')
```

```python
prices['zip'] = np.where(transformed < 25, 'Other', prices['zip'])
```

```python
prices['zip'].value_counts()
```

```python
prices.head()
```

```python
price2 = pd.get_dummies(prices.loc[:,(prices.columns!='street') & (prices.columns != 'zip') &((prices.columns != 'city'))], drop_first = True)
```

```python
price2.head()
```

```python
feature_cols = ['beds', 'baths','sq__ft','type_Multi-Family', 'type_Residential','type_Unkown']
X = price2[feature_cols]
y = price2['price']
```

```python
lr2 = LinearRegression()
```

```python
lr2.fit(X,y)
```

```python
lr2.coef_
```

```python
lr2.score(X,y)
```
